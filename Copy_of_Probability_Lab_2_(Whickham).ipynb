```python
data = {
    "age_group": ["18-44"] * 4 + ["45-64"] * 4 + ["65+"] * 4,
    "smoker": ["No", "Yes", "No", "Yes"] * 3,
    "outcome": ["Alive", "Alive", "Dead", "Dead"] * 3,
    "count": [90, 85, 10, 15, 70, 50, 30, 50, 40, 30, 60, 70]
}
df = pd.DataFrame(data)

# Normalize proportions within each age group and smoking status
df["proportion"] = df.groupby(["age_group", "smoker"])["count"].transform(lambda x: x / x.sum())

# Pivot the data to make it suitable for stacking
df_pivot = df.pivot_table(index=["age_group", "smoker"], columns="outcome", values="proportion", fill_value=0)

# Ensure "Dead" is at the bottom and "Alive" is on top
df_pivot = df_pivot[["Dead", "Alive"]]

# Plot the stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))
df_pivot.plot(kind="bar", stacked=True, ax=ax, color=["#4DB6AC", "#E57373"])  # Dead (bottom), Alive (top)
ax.set_ylabel("Proportion")
ax.set_xlabel("Smoker")
ax.set_title("Proportion of Outcomes by Smoking Status and Age Group")
ax.legend(title="Outcome")
plt.xticks(rotation=0)

# Adjust x-axis labels
ax.set_xticklabels([f"{s}\n({a})" for a, s in df_pivot.index])
plt.show()
```

```python
# Create crosstab for both age and smoking status
survival_by_age_smoking = pd.crosstab([df['Age'], df['Smoking']], df['Survival'], normalize='index')

# Create stacked bar plot
survival_by_age_smoking.plot(kind='bar',
                           stacked=True,
                           figsize=(8, 6),
                           color=['green', 'pink'])

plt.title('Survival Status by Age and Smoking')
plt.xlabel('Age Group and Smoking Status')
plt.ylabel('Proportion')
plt.legend(title='Survival Status')
plt.ylim(0, 1)  # Set y-axis limits from 0 to 1
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
