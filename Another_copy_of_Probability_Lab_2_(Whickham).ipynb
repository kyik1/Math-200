```python
import pandas as pd
```

```python
data ='https://raw.githubusercontent.com/kmath22/kmath22/main/Whichham%201.1.csv'
df = pd.read_csv(data)
df.head()
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=df)
```

```python
# @title Smoking

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Smoking').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Create a Two Way using pd.crosstab
pd.crosstab(df['Smoking'], df['Age'])
```

```python
#Create a Two Way using pd.crosstab
pd.crosstab(df['Smoking'], df['Survival'])
```

```python
# Create a Two Way crosstab with margin (margins = True)
pd.crosstab(df['Smoking'], df['Age'], margins = True)
```

```python
# For row-wise conditional probabilities (probabilities within each row)
pd.crosstab(df['Smoking'], df['Age'], margins=True, normalize='index')
```

```python
# Create a Two Way crosstab with margin (margins = True)
pd.crosstab(df['Smoking'], df['Survival'], margins = True)
```

```python
# For row-wise conditional probabilities (probabilities within each row)
pd.crosstab(df['Smoking'], df['Survival'], margins=True, normalize='index')
```
