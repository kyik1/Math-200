```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
link = 'https://raw.githubusercontent.com/kyik1/Math-200/refs/heads/main/Car%20Values%20Over%20Ti'

car_df = pd.read_csv(link)
car_df
```

```python
#import libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#Regression Print Out (MILEAGE	CAR_PRICEE)

Y = car_df['CAR_PRICE']
X = car_df['MILEAGE']
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()
print(model.summary())
```

```python
# Scatterplot with regression line (Car_Age and CAR_PRICE)
sns.regplot(x='Car_Age', y='CAR_PRICE', data=car_df)
plt.title("Car_Age vs CAR_PRICE  with Regression Line")
plt.xlabel("Car_Age")
plt.ylabel("CAR_PRICE")
plt.show()
```
